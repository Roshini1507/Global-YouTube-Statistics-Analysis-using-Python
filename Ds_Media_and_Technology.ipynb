```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# Load the dataset
df = pd.read_csv('Global YouTube Statistics.csv', encoding='latin-1')

# Ensure that the relevant columns are numeric
df['subscribers'] = pd.to_numeric(df['subscribers'], errors='coerce')
df['video views'] = pd.to_numeric(df['video views'], errors='coerce')

# Drop rows with NaN values in the relevant columns
df = df.dropna(subset=['subscribers', 'video views'])

# Calculate the Pearson correlation coefficient
correlation, p_value = pearsonr(df['subscribers'], df['video views'])

# Display the correlation coefficient and p-value
print(f"Correlation coefficient: {correlation:.4f}")
print(f"P-value: {p_value:.4f}")

# Plotting the data
plt.figure(figsize=(10, 6))
sns.scatterplot(x='subscribers', y='video views', data=df)
plt.title('Subscribers vs Total Video Views')
plt.xlabel('Number of Subscribers')
plt.ylabel('Total Video Views')
plt.xscale('log')  # Log scale for better visualization
plt.yscale('log')  # Log scale for better visualization
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Global YouTube Statistics.csv', encoding='latin-1')

# Ensure that the relevant columns are numeric
df['lowest_yearly_earnings'] = pd.to_numeric(df['lowest_yearly_earnings'], errors='coerce')
df['highest_yearly_earnings'] = pd.to_numeric(df['highest_yearly_earnings'], errors='coerce')

# Drop rows with NaN values in the relevant columns
df = df.dropna(subset=['Youtuber', 'lowest_yearly_earnings', 'highest_yearly_earnings'])

# Calculate average yearly earnings
df['average_yearly_earnings'] = (df['lowest_yearly_earnings'] + df['highest_yearly_earnings']) / 2

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['average_yearly_earnings'].quantile(0.25)
Q3 = df['average_yearly_earnings'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['average_yearly_earnings'] < lower_bound) | (df['average_yearly_earnings'] > upper_bound)]

# Display outliers
print("Outliers in Yearly Earnings:")
print(outliers[['Youtuber', 'lowest_yearly_earnings', 'highest_yearly_earnings', 'average_yearly_earnings']])

# Plotting the distribution and highlighting outliers
plt.figure(figsize=(14, 7))
sns.boxplot(x=df['average_yearly_earnings'])
plt.title('Boxplot of Average Yearly Earnings from YouTube Channels')
plt.xlabel('Average Yearly Earnings ($)')
plt.axvline(lower_bound, color='red', linestyle='--', label='Lower Bound')
plt.axvline(upper_bound, color='green', linestyle='--', label='Upper Bound')
plt.legend()
plt.show()
```

```python
import pandas as pd
from datetime import datetime

# Load the dataset containing YouTube statistics
youtube_data = pd.read_csv('Global YouTube Statistics.csv', encoding='latin-1')

# Step 1: Convert created_date to datetime format
youtube_data['created_date'] = pd.to_datetime(youtube_data['created_date'], errors='coerce')

# Step 2: Get the current date
current_date = datetime.now()

# Step 3: Calculate the number of months since the channel was created
youtube_data['months_since_creation'] = (
    (current_date - youtube_data['created_date']) / pd.Timedelta(days=30)
).astype(int)

# Step 4: Avoid division by zero for channels created in the current month
youtube_data['average_subscribers_per_month'] = youtube_data.apply(
    lambda row: row['subscribers'] / row['months_since_creation'] if row['months_since_creation'] > 0 else 0,
    axis=1
)

# Step 5: Calculate the overall average of average subscribers gained per month
overall_average = youtube_data['average_subscribers_per_month'].mean()

print(f"The overall average number of subscribers gained per month since the creation of YouTube channels is: {overall_average:.2f}")
```

```python
import pandas as pd
from datetime import datetime

# Load the dataset containing YouTube statistics
youtube_data = pd.read_csv('Global YouTube Statistics.csv', encoding='latin-1')

# Step 1: Convert created_date to datetime format
youtube_data['created_date'] = pd.to_datetime(youtube_data['created_date'], errors='coerce')

# Step 2: Get the current date
current_date = datetime.now()

# Step 3: Calculate the number of months since the channel was created
youtube_data['months_since_creation'] = (
    (current_date - youtube_data['created_date']) / pd.Timedelta(days=30)
)

# Convert 'months_since_creation' to integers, handling NaN values
# by replacing them with 0 before conversion
youtube_data['months_since_creation'] = youtube_data['months_since_creation'].fillna(0).astype(int)

# Step 4: Avoid division by zero for channels created in the current month
youtube_data['average_subscribers_per_month'] = youtube_data.apply(
    lambda row: row['subscribers'] / row['months_since_creation'] if row['months_since_creation'] > 0 else 0,
    axis=1
)

# Step 5: Calculate the overall average of average subscribers gained per month
overall_average = youtube_data['average_subscribers_per_month'].mean()

print(f"The overall average number of subscribers gained per month since the creation of YouTube channels is: {overall_average:.2f}")
```
